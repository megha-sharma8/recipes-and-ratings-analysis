```python
from lec_utils import *
import pandas as pd
import numpy as np
import plotly.express as px
pd.options.plotting.backend = 'plotly'
```

```python
raw_interactions = pd.read_csv("RAW_interactions.csv")
raw_interactions
```

***CLEANING RAW_INTERACTIONS***

```python
raw_interactions_copy = raw_interactions.copy()
raw_interactions_copy
```

```python
# cleaning user_id & recipe_id column
raw_interactions_copy['user_id'] = pd.to_numeric(raw_interactions_copy['user_id'])
raw_interactions_copy['recipe_id'] = pd.to_numeric(raw_interactions_copy['recipe_id'])
raw_interactions_copy['recipe_id']
```

```python
# identifying rows where the rating column is not an integer
def move_rating_to_review5(row):
    if isinstance(row['rating'], str) and not row['rating'].isdigit():
        row['review5'] = row['rating']
        row['rating'] = np.nan
    return row

raw_interactions_copy = raw_interactions_copy.apply(move_rating_to_review5, axis=1)
raw_interactions_copy.head()
```

```python
# cleaning date column
raw_interactions_copy['date'] = pd.to_datetime(raw_interactions_copy['date'])
raw_interactions_copy['date']
```

```python
# delete empty row
raw_interactions_copy = raw_interactions_copy.drop(columns=['Unnamed: 5'])
```

```python
raw_interactions_copy.head(15)
```

```python
raw_interactions_copy['recipe_id'].nunique()  # 185913
# makes a list of all unique reviews corresponding to the group by (recipes)
recipe_reviews_combined = raw_interactions_copy.groupby('recipe_id')['review5'].unique()
recipe_reviews_combined
```

```python
raw_interactions_copy.head(15)
```

***CLEANING RAW_RECIPES***

```python
raw_recipes = pd.read_csv("RAW_recipes.csv")
raw_recipes
```

```python
raw_recipes_copy = raw_recipes.copy()
raw_recipes_copy
```

```python
raw_recipes['id'].nunique
```

```python
raw_recipes['nutrition']
```

```python
nutrition_info = ['calories', 'total fat', 'sugar', 'sodium', 'protein', 'saturated fat', 'carbohydrates']

for i, key in enumerate(nutrition_info):
    raw_recipes[key] = raw_recipes['nutrition'].apply(lambda x: x.split(',')[i][1:])

    if i == len(nutrition_info) - 1:
        raw_recipes[key] = raw_recipes[key].apply(lambda x: x[:-1])

    raw_recipes[key] = raw_recipes[key].astype(float)

nutrition_cols = raw_recipes[['calories', 'total fat', 'sugar', 'sodium', 'protein', 'saturated fat', 'carbohydrates']]
nutrition_cols
```
